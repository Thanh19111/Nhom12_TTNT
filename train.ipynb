```python
import os

backup_dir = "/content/drive/MyDrive/human_detect"
os.makedirs(backup_dir, exist_ok=True)
```

```python
import time
import shutil

src_weights = "/content/runs/train/yolo11_run/weights/last.pt"
prev_ckpt = None
epoch = 0

print("⏳ Đang theo dõi và sao lưu từng epoch...")

while True:
    try:
        if os.path.exists(src_weights):
            # Nếu file mới hơn file trước đó, sao lưu
            if prev_ckpt is None or os.path.getmtime(src_weights) > prev_ckpt:
                epoch += 1
                dst = f"{backup_dir}/epoch_{epoch}.pt"
                shutil.copy(src_weights, dst)
                prev_ckpt = os.path.getmtime(src_weights)
                print(f"✅ Đã sao lưu epoch {epoch} vào: {dst}")
        else:
            print("⚠️ Chưa tìm thấy last.pt...")
    except Exception as e:
        print("❌ Lỗi:", e)

    time.sleep(60)  # Kiểm tra mỗi 60 giây
```
