```python
import os

backup_dir = "/content/drive/MyDrive/human_detect"
os.makedirs(backup_dir, exist_ok=True)
```

```python
import time
import shutil

src_weights = "/content/runs/train/yolo11_run/weights/last.pt"
prev_ckpt = None
epoch = 0

print("‚è≥ ƒêang theo d√µi v√† sao l∆∞u t·ª´ng epoch...")

while True:
    try:
        if os.path.exists(src_weights):
            # N·∫øu file m·ªõi h∆°n file tr∆∞·ªõc ƒë√≥, sao l∆∞u
            if prev_ckpt is None or os.path.getmtime(src_weights) > prev_ckpt:
                epoch += 1
                dst = f"{backup_dir}/epoch_{epoch}.pt"
                shutil.copy(src_weights, dst)
                prev_ckpt = os.path.getmtime(src_weights)
                print(f"‚úÖ ƒê√£ sao l∆∞u epoch {epoch} v√†o: {dst}")
        else:
            print("‚ö†Ô∏è Ch∆∞a t√¨m th·∫•y last.pt...")
    except Exception as e:
        print("‚ùå L·ªói:", e)

    time.sleep(60)  # Ki·ªÉm tra m·ªói 60 gi√¢y
```
